```python
for k in range(1, 5):
    x_values = []
    y_values = []

    for i in range(1, 6):
        x_values.append(i)
        y_values.append(i*k)

    plt.plot(x_values, y_values, label="y=" + str(k) + "x")
plt.legend()
plt.show()
```
