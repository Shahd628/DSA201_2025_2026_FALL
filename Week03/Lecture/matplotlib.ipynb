```python
f = open("iris.txt")

lines = f.readlines()

f.close()
species_to_ids = dict()
x_values = []
y_values = []
codes = []
val = 0
for line in lines[1:]:
    line = line.strip()
    info = line.split(",") # always returns a list of strings
    species = info[-1]
    sepalLength = float(info[1])
    sepalWidth = float(info[2])
    if species not in species_to_ids:
        #species_to_ids[species] = len(species_to_ids)
        species_to_ids[species] = val
        val += 1

    x_values.append(sepalLength)
    y_values.append(sepalWidth)
    codes.append(species_to_ids[species])
#print(species_to_ids)
#print(codes)
plt.scatter(x_values, y_values, c=codes)
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.show()
```

```python
f = open("iris.txt")

lines = f.readlines()

f.close()

d = {}

for line in lines[1:]:
    line = line.strip()
    info = line.split(",") # always returns a list of strings
    species = info[-1]
    sepalLength = float(info[1])
    sepalWidth = float(info[2])

    if species not in d:
        d[species] = {"x":[], "y":[]}
    
    d[species]["x"].append(sepalLength)
    d[species]["y"].append(sepalWidth)
    
for species, temp_d in d.items():
    plt.scatter(temp_d["x"], temp_d["y"], label=species)

plt.legend()
plt.xlabel("sepal Length")
plt.ylabel("sepal Width")
plt.show()
```

```python
file = open("GOOGL.txt")

lines = file.readlines()

file.close()

d = {}

for line in lines[1:]:
    line = line.strip()
    info = line.split(",")
    year = int(info[0].split("-")[0])
    close = float(info[4])
    if year not in [2004, 2022]:
        if year not in d:
            d[year] = []
        d[year].append(close)

plt.figure(figsize=(15, 10))
for year, close_values in d.items():
    plt.plot(range(1, len(close_values)+1), close_values, label=year)

plt.legend()
plt.show()
```
